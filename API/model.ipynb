```python
x=df.drop('Downtime',axis=1)
y=df['Downtime']
```

```python
plt.rcParams['figure.figsize']=[11,7]
r=3
c=4
it=1
for i in x.columns:
    plt.subplot(r,c,it)
    sns.kdeplot(x=x[i])
    plt.grid()
    it+=1
plt.tight_layout()
plt.show()
```

```python
sns.heatmap(x.corr(),annot=True)
plt.show()
```

```python
r=3
c=4
it=1
for i in x.columns:
    plt.subplot(r,c,it)
    sns.boxplot(x=x[i])
    plt.grid()
    it+=1
plt.tight_layout()
plt.show()
```

```python
y = y.replace({'Machine_Failure':1,'No_Machine_Failure':0})
y
```

```python
for i in x.columns:
    q1=x[i].quantile(0.25)
    q3=x[i].quantile(0.75)
    iqr=q3-q1
    ll=q1-(1.5*iqr)
    ul=q3+(1.5*iqr)
    for k in x[i]:
        if k<ll:
            x[i]=x[i].replace(k,ll)
        if k>ul:
            x[i]=x[i].replace(k,ul)
```

```python
x.boxplot(rot=90)
```

```python
from sklearn.preprocessing import MinMaxScaler,StandardScaler,RobustScaler
from sklearn.model_selection import train_test_split
```

```python
mmax = MinMaxScaler()
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.30,random_state=40)
```

```python
x_train.shape
```

```python
y_train.shape
```

```python
x_train = pd.DataFrame(mmax.fit_transform(x_train),columns=x_train.columns)
x_test = pd.DataFrame(mmax.fit_transform(x_test),columns=x_test.columns)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model=LogisticRegression()
model
```

```python
model.fit(x_train,y_train)
```

```python
y_pred_test=model.predict(x_test)
y_pred_train=model.predict(x_train)
```

```python
from sklearn.metrics import accuracy_score,precision_score,recall_score,f1_score
```

```python
def classification_scores(org,pred):
    print("accuracy score: ",accuracy_score(org,pred))
    print("precision score: ",precision_score(org,pred))
    print("recall score: ",recall_score(org,pred))
    print("f1 score: ",f1_score(org,pred))
classification_scores(y_train,y_pred_train)
```

```python
classification_scores(y_test,y_pred_test)
```
